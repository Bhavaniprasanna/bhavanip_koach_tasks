```python
db=client.LEVEL4
```

```python
import json
with open("input1.json") as file1:
    data1=json.load(file1)
with open("input2.json") as file2:
    data2=json.load(file2)
with open("input3.json") as file3:
    data3=json.load(file3)
#print(data1)
```

```python
#collection1
```

```python
student=db.Students
```

```python
student.insert_many(data1)
#print(result)
```

```python
#collection2
```

```python
groups=db.Groups
```

```python
groups.insert_many(data2)
```

```python
#coll3ction3
```

```python
colleges=db.Colleges
```

```python
colleges.insert_many(data3)
```

```python
#reading from the database
```

```python
fetch=student.find()
print(list(fetch))
```

```python
fetch2=colleges.find()
print(list(fetch2))
```

```python
fetching_one_record=student.find_one({'std_id': '0c5044b5-016f-4297-ae38-aa785eb5308d'})
print(fetching_one_record)
```
